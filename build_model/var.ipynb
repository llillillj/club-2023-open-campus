```python
import seaborn as sns
import matplotlib.pyplot as plt
import cv2
import os

import japanize_matplotlib
```

```python
DATA_FOLDER = "../content/gray_image_folder/"
image_names = os.listdir(DATA_FOLDER)
image_path_list = [os.path.join(DATA_FOLDER, image_name) for image_name in image_names]
```

```python
image_name = image_names[0]
fig, axs = plt.subplots(4, 3, figsize=(12, 10))
axs = axs.flatten()
for i, image_path in enumerate(image_path_list[:12]):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axs[i].imshow(image)
    axs[i].axis("off")
    axs[i].set_title(f"File name: {image_name}")
plt.suptitle("利用するデータの概観", fontsize=18)
plt.tight_layout()
plt.show()
```
